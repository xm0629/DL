```python
import os
import sys
import numpy as bp
```

```python
# 将训练的数据提取出来
def make_dataset(root):                                                         
    imgs=[]                                                                     
    n=len(os.listdir(root))//2                                                  
    for i in range(n):                                                          
        img=os.path.join(root,"%03d.png"%i)                                     
        mask=os.path.join(root,"%03d_mask.png"%i)                               
        imgs.append((img,mask))                                                 
    return imgs
    
```

```python
root = "/home/xm/Program/ALL-Data/unetdata/train"
make_dataset(root)
```
